## For train data

```python
''' as we do not need some features we will  delete them'''
df_train.drop(['Cabin','Fare','Embarked','Ticket'],axis=1,inplace=True)
```

```python
''' to get description like mean,variance,range'''
df_train.describe()
```

```python
'''now i want to make a new column having salutes of people present on titanic
   so i will create a function'''
def Salute(val):
    val1=val.split(',')[1]
    return val1.split('.')[0]
df_train["salute"]=df_train["Name"].apply(Salute)
```

```python
''' this is my new feature having salutes of all the people'''
df_train["salute"]
```

```python
''' to see how many salutes were used and 
    how many times each salute was used'''
df_train.groupby(["salute"]).count()
```

```python
''' now i will calculate meadian of ages of people having particular salute'''
median = df_train.groupby('salute')['Age'].median()
median
```

```python
''' this will give us median of ages of people having Dr as salute'''
median[" Dr"]
```

```python

'''now i am going to make a function 
   which will check if a person have nan as age,
   if yes then check its salute
   and then it will put median of ages of people of that salute
   in age of that person'''


def median_of_age(val):
    if val["salute"]==" Mr" and np.isnan(val["Age"]):
            val["Age"]=median[" Mr"]
    elif val["salute"]==" Mrs" and np.isnan(val["Age"]):
            val["Age"]=median[" Mrs"]
    elif val["salute"]==" Miss" and np.isnan(val["Age"]):
            val["Age"]=median[" Miss"]  
    elif val["salute"]==" Master" and np.isnan(val["Age"]):
            val["Age"]=median[" Master"]
    elif val["salute"]==" Dr" and np.isnan(val["Age"]):
            val["Age"]=median[" Dr"]
    elif val["salute"]==" Capt" and np.isnan(val["Age"]):
            val["Age"]=median[" Capt"]  
    elif val["salute"]==" Col" and np.isnan(val["Age"]):
            val["Age"]=median[" Col"] 
    elif val["salute"]==" Don" and np.isnan(val["Age"]):
            val["Age"]=median[" Don"] 
    elif val["salute"]==" Jonkheer" and np.isnan(val["Age"]):
            val["Age"]=median[" Jonkheer"] 
    elif val["salute"]==" Lady" and np.isnan(val["Age"]):
            val["Age"]=median[" Lady"]  
    elif val["salute"]==" Major" and np.isnan(val["Age"]):
            val["Age"]=median[" Major"]
    elif val["salute"]==" Mlle" and np.isnan(val["Age"]):
            val["Age"]=median[" Mlle"]
    elif val["salute"]==" Mme" and np.isnan(val["Age"]):
            val["Age"]=median[" Mme"] 
    elif val["salute"]==" Ms" and np.isnan(val["Age"]):
            val["Age"]=median[" Ms"]
    elif val["salute"]==" Rev" and np.isnan(val["Age"]):
             val["Age"]=median[" Rev"] 
    elif val["salute"]==" Sir" and np.isnan(val["Age"]):
            val["Age"]=median[ "Sir"]
    elif val["salute"]==" the countess" and np.isnan(val["Age"]):
            val["Age"]=median[" the countess"]   
    return val        
```

```python
''' now i am going to update my data
    by passing my function to object of my file
    and by storing it in new object'''
df_train2=df_train.apply(median_of_age,axis=1)
```

```python
''' here is my new data having updated ages'''
df_train2
```

```python
'''in my old data there were 177 missing ages'''
df_train.isnull().sum()
```

```python
'''and now i have zero missing ages'''
df_train2.isnull().sum()
```

```python
''' since i have to work on sklearn
    i will remove features stored in string format and categorical'''
df_train2.drop(["Name","salute"],axis=1,inplace=True)
```

```python
''' also will convert some features in numeric form and continous 
    as they canno be drop '''
df_train2=pd.get_dummies(df_train2)
df_train2
```

```python
''' again going to drop extra features'''
df_train2.drop('Sex_male',axis=1)
```
